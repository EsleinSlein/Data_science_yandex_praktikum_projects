## Модель

### Функция  sMAPE.

**Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных.**



- yi - Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.
- ŷi - Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.
- N - Количество объектов в выборке.

```python
def smape_score(target, predictions):
    return ((abs(target - predictions) / ((abs(target) + abs(predictions)) / 2))*100)
```

```python
#Напишем функцию для метрики
def final_smape_score(target, predictions):
    predict = pd.DataFrame(data=predictions, index=target.index)
    smape_1 = smape_score(target['rougher.output.recovery'], predict[0])
    smape_2 = smape_score(target['final.output.recovery'], predict[1])
    return np.mean(smape_1)* 0.25 + np.mean(smape_2)*0.75
```

### Обучим модели и узнаем лучшую

```python
# Создадим метрику
smape = make_scorer(final_smape_score,  greater_is_better=False)
```

```python
RANDOM_STATE=1234
```

**Так как в реальности мы не знаем эффективность обогащения и концентрацию металлов до обработки, мы избавимся от отличающихся столбцов в обучающей выборке, чтобы не было переобучения.**

```python
# Избавляемся от столбцов
train_col = df_test_1.columns

df_train_ex = df_train[train_col]

#Определяем целевые признаки
features_ex = df_train_ex.drop(['rougher.output.recovery', 'final.output.recovery', 'date'], axis=1)
target_ex = df_train_ex[['rougher.output.recovery', 'final.output.recovery']]

features_test_ex = df_test_1.drop(['rougher.output.recovery', 'final.output.recovery','date'], axis=1)
target_test_ex = df_test_1[['rougher.output.recovery', 'final.output.recovery']]
```

```python
# Масштабируем
scaler = StandardScaler()
scaler.fit(features_ex) 
features_train = scaler.transform(features_ex)
features_test = scaler.transform(features_test_ex)  
```

#### Дерево решений

```python
grid_space = { 'max_depth': [None] + [i for i in range(2, 7)],
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2),
              'max_features': ['sqrt', 'log2']
            }
```

```python
dtr = DecisionTreeRegressor(random_state= RANDOM_STATE)# обучающая модель
grid_dtr = GridSearchCV(dtr, grid_space, cv=5, scoring=smape)
grid_dtr.fit(features_ex, target_ex)

print('Best hyperparameters are: '+str(grid_dtr.best_params_))
print('Best score is: '+str(- grid_dtr.best_score_))
```

#### Случайный лес

```python
# Перебираемые гиперпараметры
grid_space_forest = { 'n_estimators': range (20, 110, 10),
              'max_features': ['sqrt', 'log2'],
              'max_depth': [None] + [i for i in range(2, 7)]
                    }
```
