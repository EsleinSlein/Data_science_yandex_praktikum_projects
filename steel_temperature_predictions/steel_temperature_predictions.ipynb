<h1> Проект по оптимизации производственных расходов в промышленности</h1> 

**Цель проекта**:

**Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Нам предстоит построить модель, которая предскажет температуру стали. Прогнозируем последнюю температуру по партии.**

**Описание этапа обработки:**

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 
Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).
Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.
Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

**Описание данных**:
Данные состоят из файлов, полученных из разных источников
- __data_arc_new.csv__ — данные об электродах;
- __data_bulk_new.csv.csv__ — данные о подаче сыпучих материалов (объём);
- __data_bulk_time_new.csv__ — данные о подаче сыпучих материалов (время);
- __data_gas_new.csv__ — данные о продувке сплава газом;
- __data_temp_new.csv__ — результаты измерения температуры;
- __data_wire_new.csv__ — данные о проволочных материалах (объём);
- __data_wire_time_new.csv__ — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

**Метрика качества модели MAE. Модель считается успешной, если MAE<6.8**

Так как активная мощность:
$$P=U*I*cosφ $$
А реактивная:
$$ Q=U*I*sinφ $$
и следовательно данные коррелируют. Мы создадим новый признак `Е`- энергия: 
$$E=S*T $$ 

где $$S=\sqrt{P^2 + Q^2}$$, $$ T = Начало\ нагрева\ дугой - Конец \ нагрева\ дугой $$

 **Создаем множество из дефектных партий `anomaly_arc`, `bulk_anomaly`, `gas_anomaly`, `empty`, `wire_anomaly` которые мы определили ранее. Также добавим отрицательное реактивную мощность -_2116_. ## Данные об электродах

**Нужно избавиться от аномалий. Привести столбцы к snake_case, перевести на английский, и создать новые признаки**

```python
#приводим столбцы в поорядок
df_arc.columns = ['key', 'time_start','time_end', 'active_power', 'reactive_power']
```

```python
#Избавляемся от аномалий
df_arc = df_arc[df_arc['key'].isin(delete_1) == False]
display(df_arc.shape)
df_arc.describe()
```

**Т.к. мощность определяется как джоуль/с, то и генерируемое время должно быть в секундах**

```python
# Создаем признак time
df_arc['time'] = df_arc['time_end'].astype('int64') - df_arc['time_start'].astype('int64')
df_arc['time'] = df_arc['time'] // 10**9
display(df_arc.head())
df_arc.shape
```

```python
#Находим полную мощность
df_arc['full_power'] = (df_arc['active_power']**2 + df_arc['reactive_power']**2)**0.5

#Создаем признак электроэнергии
df_arc['energy'] = df_arc['time'] * df_arc['full_power']
display(df_arc.head())
df_arc.shape
```

**Т.к.  `active_power`, `reactive_power`, `time`, `full_power`, `energy` линейно зависимы мы оставим в признаках только энергию.**

```python
# Группируем по ключу и избавляемся от лишних признаков
df_arc = df_arc.groupby('key').agg(energy = ('energy','sum'))
display(df_arc.head())
df_arc.info()
```

**Разобьем данные на выборки. Обучим 3 модели и с памощью кросс-валидации отберем лучшую. Проверим на тестовой выборке и определим "важные" признаки. Такие признаки больше всего оказывают влияние на обобщаемость модели. Сравним лучшую модель с константной моделью.**

## Подготовка данных для обучения

```python
#Разобьем данные на выборки
features = df_finall.drop('temp_last', axis=1)
target = df_finall['temp_last']

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE
)
```

##  Модель Ridge 

```python
model_ridge = Ridge(random_state=RANDOM_STATE)

ridge_pipeline = make_pipeline(StandardScaler(), model_ridge)

param_grid_ridge = {
    'ridge__alpha': np.arange(0, 1, 0.01),
}
```

```python
rs_ridge_pl = RandomizedSearchCV(
    ridge_pipeline,
    param_grid_ridge,
    cv=3,
    scoring='neg_mean_absolute_error',
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=1
)

rs_ridge_pl.fit(features_train, target_train)

# лучшее значение MAE на кросс-валидации
print(f'best_score: {-rs_ridge_pl.best_score_}')
# лучшие гиперпараметры
print(f'best_params: {rs_ridge_pl.best_estimator_}')
```

## Модель RandomForest

```python
model_rf = RandomForestRegressor(random_state=RANDOM_STATE)

param_grid_rf = {
    'n_estimators': range(100, 800, 100),
    'max_depth':range(2, 11),
    'min_samples_split': (2, 3, 4),
}
```

```python
rs_forest = RandomizedSearchCV(
    model_rf,
    param_grid_rf,
    cv=3,
    scoring='neg_mean_absolute_error',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

rs_forest.fit(features_train, target_train)

# лучшее значение MAE на кросс-валидации
print(f'best_score: {-rs_forest.best_score_}')
# лучшие гиперпараметры
print(f'best_params: {rs_forest.best_estimator_}')
```

## Модель CatBoost

```python
model_cat = CatBoostRegressor(early_stopping_rounds=50, random_state=RANDOM_STATE,  verbose=False)

param_grid_cat = {
    'loss_function': ['RMSE', 'MAE'],
    'max_depth': range(2, 11),
    'bootstrap_type': ['Bayesian', 'Bernoulli'],
    'n_estimators': range(100, 900, 100),
}
```

```python
rs_cat = RandomizedSearchCV(
    model_cat,
    param_grid_cat,
    cv=3,
    scoring='neg_mean_absolute_error',
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=1
)

rs_cat.fit(features_train, target_train)

# лучшее значение MAE на кросс-валидации
print(f'best_score: {-rs_cat.best_score_}')
# лучшие гиперпараметры
print(f'best_params: {rs_cat.best_params_}')
```

**Вывод**: CatBoost дает лучший результат. Его и проверим на тестовой.
